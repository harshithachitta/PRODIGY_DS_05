```python
df.drop(['Precipitation(in)','Weather_Condition','Wind_Speed(mph)', 'Wind_Chill(F)','Weather_Timestamp'],
        axis = 1, inplace = True)
df.head()
```

```python
categorical=[i for i in df.columns if df[i].dtype=='O']
print('The categorical variables are',categorical)
```

```python
for i in categorical:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
plt.figure(figsize=(10,10))
plotnumber = 1
for i in numerical:
    if plotnumber <= df.shape[1]:
        ax1 = plt.subplot(2,2,plotnumber)
        plt.hist(df[i],color='red')
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.title('frequency of '+i, fontsize=10)
    plotnumber +=1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

#creating a new data frame from performing the chi2 analysis
df1=pd.DataFrame()

#adding all the categorical columns except the output to new data frame
for i in categorical:
    if i!= 'Accident_severity':
        df1[i]=le.fit_transform(df[i])
```

```python
import pandas as pd

# Assuming 'df1' contains your categorical features and 'df' contains 'Visibility(mi)'
# Select the numerical features from 'df1' (if any) and combine with 'Visibility(mi)'
numerical_features = df1.select_dtypes(include=['number'])
data_for_correlation = pd.concat([numerical_features, df['Visibility(mi)']], axis=1)

# Calculate the correlation matrix
correlation_matrix = data_for_correlation.corr()

# Print the correlation between features and 'Visibility(mi)'
print(correlation_matrix['Visibility(mi)'])
```

```python
import pandas as pd

# Assuming 'df1' contains your categorical features and 'df' contains 'Visibility(mi)'
# Select the numerical features from 'df1' and combine with 'Visibility(mi)'
numerical_features = df1.select_dtypes(include=['number'])
data_for_correlation = pd.concat([numerical_features, df['Visibility(mi)']], axis=1)

# Calculate the correlation matrix
correlation_matrix = data_for_correlation.corr()

# Get the correlation values for 'Visibility(mi)' with other features
f_p_values = correlation_matrix['Visibility(mi)']

# Print the correlation values (f_p_values)
print(f_p_values)
```

```python
f_p_values1 = pd.DataFrame({'features': numerical_features.columns, 'Correlation': f_p_values[:-1]})  # Exclude 'Visibility(mi)' itself
print(f_p_values1)
```

```python
f_p_values1.sort_values(by='Correlation', ascending=False, inplace=True)
print(f_p_values1)
```
